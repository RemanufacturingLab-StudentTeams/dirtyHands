```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from scipy.spatial import ConvexHull
import pandas as pd
import random

# Change to your csv path
PATH = 'hypergraph_test_form_data.csv'

# Function to generate offset points
def generate_offset_points(center, num_points=60, radius=0.05):
    """ Generate offset points around a center point. """
    angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)
    return np.array([center[0] + radius * np.cos(angles), center[1] + radius * np.sin(angles)]).T

# Function to plot convex hulls
def plot_convex_hull(ax, points, color):
    """ Plot a convex hull for given points with a specified color. """
    if len(points) < 3:  # Convex hull requires at least 3 points
        return
    hull = ConvexHull(points)
    polygon = plt.Polygon(points[hull.vertices], closed=True, color=color, alpha=0.3)
    ax.add_patch(polygon)

# Load the CSV file when using semicolons as separators
df = pd.read_csv(PATH, sep=';')

# Load the CSV file when using commas as separators
# df = pd.read_csv(PATH)

# Initialize the graph
G = nx.Graph()

# Process connections from the data
for idx, connection in df.iterrows():
    parent_id = str(connection['parent_part'])
    child_id = str(connection['id_number'])  # Convert to string for consistency

    if parent_id != 'none':
        G.add_edge(parent_id, child_id)  # Add an edge
        edge_label = f"Accessibility: {connection['accessibility']}\nType: {connection['part_name']}"
        G.add_edge(parent_id, child_id, label=edge_label)
    else:
        G.add_node(child_id)  # Add a standalone node

# Create positions using NetworkX's spring layout
pos = nx.spring_layout(G, seed=42)

# Plot the graph
plt.figure(figsize=(12, 10))
nx.draw(
    G, pos,
    with_labels=True,
    node_color='lightblue',
    edge_color='gray',
    node_size=500,
    alpha=1
)



edge_labels = {(u, v): f"C{i}: ({df.at[int(v)-1, 'connection_type']},{df.at[int(v)-1, 'accessibility']})" 
               for i, (u, v) in enumerate(G.edges, start=0)}

# Draw edge labels on the graph
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color='red',
    font_size=12
)

parent_ids = set(df['parent_part'].dropna().astype(str))  # Get unique Parent IDs from the DataFrame

# Create labels for parent nodes only
parent_labels = {node: f"Parent {node}" for node in parent_ids if node in G.nodes()}

# Add labels to parent nodes
for node, (x, y) in pos.items():
    if node in parent_ids:
        assembly_label = df.at[int(node)-1, 'part_name']
        plt.text(
            x, y-0.1,  # Adjust position slightly above the node
            f"{assembly_label}",
            fontsize=18,
            color='black',
            ha='center', va='center',
            bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightyellow', alpha=0.8)
        )


# Group nodes by parent ID
parent_groups = {}
for idx, connection in df.iterrows():
    parent_id = str(connection['parent_part'])
    child_id = str(connection['id_number'])
    
    if parent_id != 'none':
        if parent_id not in parent_groups:
            parent_groups[parent_id] = []
        parent_groups[parent_id].append(child_id)
        parent_groups[parent_id].append(parent_id)

# Track handles for legend
legend_handles = []

# Iterate through each group to calculate and plot the convex hulls

assembly_id = 1

for parent, nodes in parent_groups.items():
    unique_nodes = list(set(nodes))  # Ensure unique nodes

    # Get the positions of the nodes in the group
    valid_positions = [pos[node] for node in unique_nodes if node in pos]

    if not valid_positions:
        continue  # Skip if no valid positions are found

    # Generate offset points for the valid positions
    offset_points = np.vstack([generate_offset_points(point) for point in valid_positions])

    # Check if there are enough points to form a convex hull
    if len(offset_points) >= 2:
        color = np.random.rand(3,)  # Generate a random color
        plot_convex_hull(plt.gca(), offset_points, color)

        # Add legend handle
        legend_handles.append(plt.Line2D([0], [0], marker='o', color='w', label=f'Assembly {assembly_id}', 
                                           markerfacecolor=color, markersize=10))
    
    assembly_id += 1

# Add a legend
plt.legend(handles=legend_handles, loc='upper left', bbox_to_anchor=(1, 1), title="Assemblies")

# Add a title based on product name
plt.title(df.loc[df['id_number'] == 1, 'brand_name'].iloc[0] + " " + df.loc[df['id_number'] == 1, 'model'].iloc[0])
plt.axis('equal')  # Equal aspect ratio for the plot

plt.savefig("fairphone_2_graphSHORTEST.pdf", format="pdf", bbox_inches="tight")

plt.show()

plt.show()
```
